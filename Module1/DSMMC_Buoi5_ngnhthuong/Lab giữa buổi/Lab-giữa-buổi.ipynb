# Problem 1: Tạo hai ma trận 3x3 với các phần tử integer ngẫu nhiên từ 1 - 10.

```python
# Your code here
import numpy as np
matran1 = np.random.randint(1,11,size=[3,3])
matran2 = np.random.randint(1,11,size=[3,3])
print ('Ma trận 1 là: ', matran1) #Đăt đây để k dính for sẽ bị lặp lại
print ('Ma trận 2 là: ', matran2)
```

# Problem 2: Thực hiện phép cộng phần tử, nhân từng phần tử, tích vô hướng với hai ma trận vừa được tạo.


```python
# Your code here
print('cộng 2 ma trận: ', matran1+matran2)
print('Nhân 2 ma trận: ', matran1*matran2)
print('Tích vô hướng của 2 ma trận: ', np.dot(matran1,matran2))
```
